```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
x1_data = pd.read_csv('C:/Users/.../x1.txt', header=None, sep=',').dropna(axis=1).values
x2_data = pd.read_csv('C:/Users/.../x2.txt', header=None, sep=',').dropna(axis=1).values
pts_data = pd.read_csv('C:/Users/.../pts.txt', header=None, sep=',').dropna(axis=1).values
```

```python
def report_package_params(dataset_name, model):
    print(f'{dataset_name} Weights: \n{model.weights_}\n')
    print(f'{dataset_name} Means: \n{model.means_}\n')
    print(f'{dataset_name} Covariances: \n{model.covariances_}\n')
```

Implement the EM algorithm for clustering with multivariate Gaussian models discussed in the LearningGM section. Initialize the algorithm with the K-means result. Plot the obtained clusters in different colors.

```python
x1_GM = GaussianMixture(n_components=2, init_params='kmeans')
x1_GM.fit(x1_data)
report_package_params("x1", x1_GM)
```

```python
x2_GM = GaussianMixture(n_components=2, init_params='kmeans')
x2_GM.fit(x2_data)
report_package_params("x2", x2_GM)
```

```python
pts_GM = GaussianMixture(n_components=3, init_params='kmeans')
pts_GM.fit(pts_data)
report_package_params("pts", pts_GM)
```

```python
labels1 = x1_GM.predict(x1_data)

plt.scatter(x1_data[:, 0], x1_data[:, 1], c=labels1, s=40, cmap='viridis')
```

```python
labels2 = x2_GM.predict(x2_data)

plt.scatter(x2_data[:, 0], x2_data[:, 1], c=labels2, s=40, cmap='viridis')
```

```python
labels3 = pts_GM.predict(pts_data)

plt.scatter(pts_data[:, 0], pts_data[:, 1], c=labels3, s=40, cmap='viridis')
```

Implement the spectral clustering algorithm and repeat the questions.

```python
from sklearn.cluster import SpectralClustering

clustering_x1 = SpectralClustering(n_clusters=2,random_state=0,assign_labels='kmeans',
                                   affinity='nearest_neighbors').fit(x1_data)

plt.scatter(x1_data[:,0],x1_data[:,1], c=clustering_x1.labels_)
```

```python
clustering_x2 = SpectralClustering(n_clusters=2,random_state=0,assign_labels='kmeans',
                                   affinity='nearest_neighbors').fit(x2_data)

plt.scatter(x2_data[:,0],x2_data[:,1], c=clustering_x2.labels_)
```

```python
clustering_pts = SpectralClustering(n_clusters=3,random_state=0,assign_labels='kmeans',
                                    affinity='nearest_neighbors').fit(pts_data)

plt.scatter(pts_data[:,0],pts_data[:,1], c=clustering_pts.labels_)
```
